```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
gp1 = ('mhead2', 'mhead2')
gp2 = ('amide', 'oh1')
gps = [gp1, gp2]
ylims = [-10, 10]
legend = []

for gp in gps:
    if os.path.exists('{}-{}.txt'.format(gp[0], gp[1])):
        filename = '{}-{}.txt'.format(gp[0], gp[1])
        dat = np.loadtxt(filename)
    elif os.path.exists('{}-{}.txt'.format(gp[1], gp[0])):
        filename = '{}-{}.txt'.format(gp[1], gp[0])
        dat = np.loadtxt(filename)
    else:
        print('No such file exists')
    legend.append(filename)
        
    plt.plot(dat[:,0], dat[:,1])
    plt.ylim(ylims)
plt.legend(legend)
```
